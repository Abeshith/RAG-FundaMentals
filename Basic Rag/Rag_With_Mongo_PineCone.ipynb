```python
from datasets import load_dataset
import pandas as pd

dataset = load_dataset("MongoDB/embedded_movies")
```

```python
df = pd.DataFrame(dataset['train'])
df.shape
```

```python
data = df.sample(100)
data.head(2)
```

```python
data.isnull().sum()
```

```python
data = data.dropna(subset=['fullplot'])
data = data.drop(columns=['plot_embedding'])
```

```python
import pymongo
from pymongo.mongo_client import MongoClient
from google.colab import userdata

uri = userdata.get("MONGODB_URI")

client = MongoClient(uri)

try:
    client.admin.command('ping')
    print("Pinged your deployment. You successfully connected to MongoDB!")
except Exception as e:
    print(e)
```

```python
db = client["mongo_pinecone_rag"]
collection = db["mongo_pinecone_rag_collection"]
```

```python
document = data.to_dict(orient='records')
```

```python
collection.insert_many(document)
```

```python
from pinecone import Pinecone

pinecone = Pinecone(api_key= userdata.get('PINECONE_API_KEY'))
index = pinecone.Index("rag")
```

```python
from sentence_transformers import SentenceTransformer
embedding_model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
```

```python
def get_result(query, similiar_result = 4):
  embedding = embedding_model.encode(query)
  embedding = embedding.tolist()

  result = index.query(vector=embedding, top_k=similiar_result, include_metadata= True)
  return result
```

```python
query = "What is the Best Horror Movie to Watch andd Why?"
result = get_result(query)
print(result)
```

```python
from bson.objectid import ObjectId

mylist = []
for i in range(len(result["matches"])):
  value = result["matches"][i]['id']
  mylist.append(collection.find_one({"_id": ObjectId(value)}))
```

```python
mylist
```

```python
combined_information = ""

for i in range(len(mylist)):
    fullplot = mylist[i]["fullplot"]
    title = mylist[i]["title"]
    combined_information += f"Title: {title}, Plot: {fullplot}\n"

print(combined_information)
```

```python
prompt = f"Query: {query}\n Continue to answer the query by using the fullplot only: \n{combined_information}"
print(prompt)
```

```python
from langchain_groq import ChatGroq

llm = ChatGroq(model = "gemma2-9b-it", api_key= userdata.get("GROQ_API_KEY"))
```

```python
response = llm.invoke(prompt).content

print(response)
```
