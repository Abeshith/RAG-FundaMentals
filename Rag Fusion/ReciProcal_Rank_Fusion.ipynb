```python
import textwrap

def wrap_text(text, width=90):
  lines = text.split("\n")

  wrapped_lines = [textwrap.fill(line, width=width) for line in lines]

  wrapped_text = '\n'.join(wrapped_lines)

  return wrapped_text
```

```python
from langchain.document_loaders import TextLoader
from langchain.document_loaders import DirectoryLoader
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
data_path = "/content/drive/MyDrive/Classroom/SingaporeTouristPlaces"
```

```python
%%time
loader = DirectoryLoader(data_path, glob = "*.txt", show_progress = True)
docs = loader.load()
```

```python
docs[0]
```

```python
raw_text = ''

for i, doc in enumerate(docs):
  text = doc.page_content

  if text:
    raw_text += text
```

```python
raw_text
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 500,
    chunk_overlap = 100,
    length_function = len,
    is_separator_regex = False,
)
```

```python
texts = text_splitter.split_text(raw_text)
len(texts)
```

```python
texts[5]
```

```python
from langchain.embeddings import HuggingFaceBgeEmbeddings
```

```python
embedding = HuggingFaceBgeEmbeddings(
    model_name = "BAAI/bge-small-en-v1.5",
    model_kwargs = {'device': 'cuda'},
    encode_kwargs = {'normalize_embeddings': True}
)
```

```python
from langchain.vectorstores import Chroma
```

```python
vector_store = Chroma.from_texts(
    texts,
    embedding,
    persist_directory = "./db"
)
```

```python
query = "Tell me About the Safari in Singapore"
vector_store.similarity_search(query , k = 5)
```

```python
retriever = vector_store.as_retriever(search_type = "mmr", k = 3)

retriever.get_relevant_documents(query)
```

```python
from operator import itemgetter
from langchain.prompts import ChatPromptTemplate
from langchain.schema.output_parser import StrOutputParser
from langchain.schema.runnable import RunnableLambda, RunnablePassthrough
```

```python
template = """ Answer the Question Based Only on the Following Context:
{context}

Question: {question}
"""
```

```python
prompt = ChatPromptTemplate.from_template(template)
prompt
```

```python
chain = (
    {"context": retriever, "question": RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)
```

```python
response = chain.invoke("Tell Me About The Universal Studios Singapore")

print(wrap_text(response))
```

# Rag Fusion

```python
from langchain.prompts import SystemMessagePromptTemplate, HumanMessagePromptTemplate
from langchain.schema.output_parser import StrOutputParser
from langchain.prompts import PromptTemplate, ChatMessagePromptTemplate, ChatPromptTemplate
```

```python
prompt = ChatPromptTemplate(
    input_variables = ["original_query"],
    messages = [SystemMessagePromptTemplate(prompt = PromptTemplate(input_variables = [] , template = "You are a helpful Assistant that generates multiple search queries based on a single input query.")),
                HumanMessagePromptTemplate(prompt = PromptTemplate(input_variables = ["original_query"], template = "Generate multiple search queries related to :{question} \n OUTPUT (4 queries):"))]
)
```

```python
original_query = "Universal Studios Singapore"
```

```python
generate_queries = (
    prompt | llm | StrOutputParser() | (lambda x : x.split("\n"))
)

generate_queries
```

```python
from langchain.load import dumps, loads

def reciprocal_rank_fusion(results: list[list], k=60):
    fused_scores = {}
    for docs in results:
        # Assumes the docs are returned in sorted order of relevance
        for rank, doc in enumerate(docs):
            doc_str = dumps(doc)
            if doc_str not in fused_scores:
                fused_scores[doc_str] = 0
            previous_score = fused_scores[doc_str]
            fused_scores[doc_str] += 1 / (rank + k)

    reranked_results = [
        (loads(doc), score)
        for doc, score in sorted(fused_scores.items(), key=lambda x: x[1], reverse=True)
    ]
    return reranked_results
```

```python
ragfusion_chain = generate_queries | retriever.map() | reciprocal_rank_fusion
```

```python
import langchain
langchain.debug = True
```

```python
ragfusion_chain.input_schema.schema()
```

```python
ragfusion_chain.invoke({"question": original_query})
```

```python
from langchain.schema.runnable import RunnablePassthrough
template = """Answer the question based only on the following context:
{context}

Question: {question}
"""
prompt = ChatPromptTemplate.from_template(template)

full_rag_fusion_chain = (
    {
        "context": ragfusion_chain,
        "question": RunnablePassthrough()
    }
    | prompt
    | llm
    | StrOutputParser()
)
```

```python
full_rag_fusion_chain.invoke({"question": "Tell me About the Sentosa Island of Singapore"})
```
